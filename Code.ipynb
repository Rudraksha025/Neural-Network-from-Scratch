```python
def init_params():
    w1 = np.random.randn(10, x_train.shape[0]) * 0.01
    b1 = np.zeros((10, 1))
    w2 = np.random.randn(10, 10) * 0.01
    b2 = np.zeros((10, 1))
    return w1, b1, w2, b2
```

```python
def Relu(z):
    return np.maximum(0,z)
```

```python
#def softmax(z):
#    A = np.exp(z) / np.sum(np.exp(z))
#    return A
def softmax(z):
    z -= np.max(z, axis=0, keepdims=True)  # stability trick
    exp_z = np.exp(z)
    return exp_z / np.sum(exp_z, axis=0, keepdims=True)
```

```python
def forward_prop(w1,b1,w2,b2,x):
    z1 = w1.dot(x) + b1
    a1 = Relu(z1)
    z2 = w2.dot(a1) + b2
    a2 = softmax(z2)
    return z1,a1,z2,a2
```

```python
def one_hot(y):
    one_hot_y = np.zeros((y.size, y.max()+1))
    one_hot_y[(np.arange(y.size), y)] = 1
    one_hot_y = one_hot_y.T
    return one_hot_y
```

```python
def deriv_relu(z):
    return z>0
```

```python
def back_prop(z1,a1,z2,a2,w2,x,y):
    m = y.size
    one_hot_y = one_hot(y)
    dz2 = a2 - one_hot_y
    dw2 = 1/m * dz2.dot(a1.T)
    db2 = 1/m * np.sum(dz2, axis=1, keepdims=True)
    dz1 = w2.T.dot(dz2) * deriv_relu(z1)
    dw1 = 1/m * dz1.dot(x.T)
    db1 = 1/m * np.sum(dz1, axis=1, keepdims=True) 
    return dw1,db1,dw2,db2
```

```python
def update_params(w1, b1, w2, b2, dw1, db1, dw2, db2, alpha):
    w1 = w1 - alpha * dw1
    b1 = b1 - alpha * db1
    w2 = w2 - alpha * dw2
    b2 = b2 - alpha * db2
    return w1,b1,w2,b2
```

```python
def get_predictions(A2):
    return np.argmax(A2,0)
```

```python
def get_accuracy(predictions, y):
    print(predictions, y)
    return np.sum(predictions == y)/y.size
```

```python
def gradient_descent(x, y, iterations, alpha):
    w1,b1,w2,b2 = init_params()
    for i in range(iterations):
        z1,a1,z2,a2 = forward_prop(w1,b1,w2,b2,x)

        dw1,db1,dw2,db2 = back_prop(z1,a1,z2,a2,w2,x,y)

        w1,b1,w2,b2 = update_params(w1,b1,w2,b2,dw1,db1,dw2,db2,alpha)

        if (i % 10 == 0 and i>0):
            print("iteration: ", i) 
            print("Accuracy: ", get_accuracy(get_predictions(a2), y))
    return w1,b1,w2,b2
```

```python
w1,b1,w2,b2 = gradient_descent(x_train, y_train, 500, 0.05)
```

```python
# Forward pass on dev (test) set
_, _, _, a2_dev = forward_prop(w1, b1, w2, b2, x_dev)

# Accuracy on dev set
print("Test Accuracy:", get_accuracy(get_predictions(a2_dev), y_dev))
```
